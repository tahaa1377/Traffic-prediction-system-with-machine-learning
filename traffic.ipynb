```python
import pandas as pd
import numpy as np

df = pd.read_excel(r'/content/data.xlsx', engine='openpyxl',skiprows = 1)
df.drop(['نام محور', 'کد محور','تعداد کل وسیله نقلیه','تعداد وسیله نقلیه کلاس 1','تعداد تخلف سبقت غیر مجاز'], axis=1, inplace=True)
df.drop(['تعداد وسیله نقلیه کلاس 2','تعداد وسیله نقلیه کلاس 3','تعداد وسیله نقلیه کلاس 4','تعداد وسیله نقلیه کلاس 5','مدت زمان کارکرد (دقیقه)'], axis=1, inplace=True)
df
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]


train_x = np.array(train[["روز","ساعت","سرعت متوسط","تعداد تخلف سرعت غیر مجاز","تعداد تخلف فاصله غیر مجاز"]])
train_y = np.array(train[['تعداد برآورد شده']])
reg = LinearRegression().fit(train_x, train_y)
print("score : %.3f" % reg.score(train_x, train_y))
print(reg.coef_[0])
print("intercept : %.3f" % reg.intercept_[0])


#----------------------------------------------------------------
test_x = np.array(test[["روز","ساعت","سرعت متوسط","تعداد تخلف سرعت غیر مجاز","تعداد تخلف فاصله غیر مجاز"]])
test_y = np.array(test[['تعداد برآورد شده']])
test_y_ = reg.predict(test_x)
print("r2 score : %.2f" % (r2_score(test_y,test_y_)))
```
